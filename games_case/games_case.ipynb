### Обработка пропусков в столбце в рейтингом

```python
df['rating'].unique()
```

Наконец, столбец с рейтингом от ESRB. Я, если честно, не знаком с рейтингами игр и их расшифровками. Полез узнавать. 

EC - early childhood (3+)

E - everyone (0+)

E10+ - everyone 10 and older (10+) 

T - teen (13+) 

M - mature (17+)

AO - adults only (18+)

RP - rating pending (рейтинг не установлен, ожидается). 

K-A - kids to adults (6+) - это устаревший рейтинг. Он сейчас в системе оценок ESRB заменен на E. Заменю его и я в соответствующем столбце. 

```python
df['rating'] = df['rating'].replace('K-A','E')
df['rating'].unique()
```

```python
df[df['rating'].isna()]
```

Пропуски с рейтингом можно частично заполнить. Одна и та же игра выходит на нескольких платформах и записана в нескольких строках нашего датафрейма. Для одних из них рейтинг будет проставлен. Здесь можно утверждать, что на разных консолях будет указан один и тот же рейтинг, так как он (рейтинг) не привязан непосредственно к консоли, а привязан именно в игре. 

```python
df['rating'] = df['rating'].fillna(df.groupby('name')['rating'].transform('first'))
df.loc[df['rating'].isna()]
```

Ну, чуть-чуть восстановили данные :) С остальными пропусками, опять же, не вижу смысла что-либо делать, поскольку заполнение исказит данные сильнее, чем просто пропуски. 

Заменю также пропуски в столбце с рейтингом на `No rating` для дальнейшего анализа в работе. Так будет удобнее оценивать влияние рейтинга(или его отсутствие) на продажи игр.

```python
df['rating'] = df['rating'].fillna('No rating')
df['rating'].unique()
```

Теперь у нас нет пропусков в столбце с рейтингом.

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Поскольку мы будем сравнивать средние двух совокупностей, то нам необходимо обращать внимание как на сами средние, так и на дисперсии, потому что разброс данных в совокупностях может повлиять на результат статистического теста. 

В качестве нулевой гипотезу я возьму следующую: 
`H₀ - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые`.
В качестве альтернативной: 
`H₀ - Средние пользовательские рейтинги платформ Xbox One и PC отличаются`. Альтернативная гипотеза двусторонняя. 

В качестве уровня значимости я возьму значение в 5%.
Чтобы тест прошел корректно, я уберу пустые строки в столбце `user_score` в обоих сравниваемых выборках и далее проведу сам тест. 

Генеральные совокупности независимы друг от друга. В выборке больше 30 значений. Построим график распределения значений по столбцу `user_score` для каждой платформы и посчитаем средние и дисперсии перед проведением теста. 

```python
XOne = data[data['platform'] == 'XOne']
XOne = XOne.dropna(subset=['user_score'])
PC = data[data['platform'] == 'PC']
PC = PC.dropna(subset=['user_score'])
```

```python
ax = XOne['user_score'].hist(bins=25, figsize=(10,6), label='Распределение оценок платформы XOne', alpha=1)
PC['user_score'].hist(bins=25, figsize=(10,6), label='Распределение оценок платформы PC', ax=ax, alpha=0.3)
plt.title('Распределение данных оценок пользователей обеих платформ')
plt.ylabel('Значения')
plt.xlabel('Шкала оценок')
plt.show()
```

```python
XOne_estimate = np.var(XOne['user_score'])
PC_estimate = np.var(PC['user_score'])
XOne_std = np.std(XOne['user_score'])
PC_std = np.std(PC ['user_score'])
print('Дисперсия оценок на платформе XOne:', XOne_estimate)
print('Дисперсия оценок на платформе PC:', PC_estimate)
print(' ')
print('Стандартное отклонение XOne:', XOne_std)
print('Стандартное отклонение PC:', PC_std)
```

```python
print('Среднее значение оценки пользователей платформы XOne:', XOne['user_score'].mean())
print('Среднее значение оценки пользователей платформы PC:', PC['user_score'].mean())
```

Распределение смещенное вправо. При этом, частота значений больше "5" сильно выше, чем ниже этого значения. Дисперсии двух выборок сильно отличаются друг от друга: дисперсия значений выборки PC сильно выше, чем в выборе XOne. Это говорит о том, что в данных по платформе PC сильнее разбросаны относительно данных по платформе XOne. Также это видно по стандартному отклонению. 

```python
alpha = .05

result = st.ttest_ind(XOne['user_score'], PC['user_score'], equal_var=False) 
print('p-значение: ', result.pvalue)
```

### Результаты теста

Полученное значение p-value говорит о том, что вероятность получить такие же или более значимые различия равна почти 29%. Это очень большое значение, чтобы делать вывод, что средние оценки пользователей двух платформ разные. 
Более того, если взглянем на сами средние значения, мы увидим, что они практически одинаковые.

Таким образом, у нас мало оснований отвергать нашу нулевую гипотезу. 

### Тест Бартлетта

```python
stat, p = bartlett(XOne['user_score'], PC['user_score'])
p
```

В качестве альтернативной: "Рассматриваемые выборки получены из генеральных совокупностей, обладающих отличающимися дисперсиями".
